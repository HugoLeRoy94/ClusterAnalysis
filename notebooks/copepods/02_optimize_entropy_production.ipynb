```python
df1 = pd.read_csv('../../data/copepods/interim/entropy_scan.csv')
#df2 = pd.read_csv('../data/interim/entropy_scan_2.csv')
#df = pd.concat([df1,df2],ignore_index=True)
df = df1
```

Look at the entropy production using absolute value of the torsion angle.

```python
df1_abs = pd.read_csv('../../data/copepods/interim/entropy_scan_w_torsion.csv')
#df2_abs = pd.read_csv('../data/interim/entropy_scan_abs_torsion.csv')
#df_abs = pd.concat([df1_abs,df2_abs],ignore_index=True)
df_abs = df1_abs
```

```python
fig,ax = plt.subplots(ncols=2,figsize=(14,7))

for K, group in df.groupby("K"):
    ax[0].plot(group["n_clusters"], group["entropy_rate"], marker='o', label=f"K={K}")

ax[0].set_xlabel("Number of clusters ")
ax[0].set_ylabel("Entropy rate (bits)")
ax[0].set_title("Entropy rate vs number of clusters")
#ax[0].set_xlim(0,5)
ax[0].legend()

for K, group in df_abs.groupby("K"):
    ax[1].plot(group["n_clusters"], group["entropy_rate"], marker='o', label=f"K={K}")

ax[1].set_xlabel("Number of clusters ")
ax[1].set_ylabel("Entropy rate (bits)")
ax[1].set_title("Entropy rate vs number of clusters")
#ax[1].set_xlim(0,5)
ax[1].legend()
```
